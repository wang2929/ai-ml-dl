# Text Pre-processing
Demoing text preprocessing with ./data/story.txt

```python
story = None
with open("./data/story.txt", 'r') as file:
    story = file.read()
```

Text pre-processing includes removing noise, tokenization, and normalization.

### Removing Noise
Removing some noise here. Noise includes capital letters, special characters, and extra white spaces.

```python
import re
def removing_noise(txt_file:str) -> str:
    # flatten the string
    txt_file = txt_file.lower()
    # Remove markdown symbols explicitly
    txt_file = re.sub(r'[#*_>`~\-]', ' ', txt_file)
    # removing special characters
    txt_file = re.sub(r'[^a-z]\s|\.', ' ', txt_file)
    # normalize whitespaces
    txt_file = re.sub(r'\s+', ' ', txt_file)
    return txt_file
```

### Tokenization
Adding tokenizers from NLTK library.

```python
import nltk
nltk.download('punkt')      # Tokenizer models returns True
nltk.download('punkt_tab')  # brings in other tokenizer methods
nltk.download('stopwords')  # List of common stopwords returns True
nltk.download('wordnet')    # For lemmatization returns True
nltk.download('averaged_perceptron_tagger')  # For POS tagging returns True
nltk.download('averaged_perceptron_tagger_eng')
```

This tokenizaiton breaks up the string into a list of sentences. First tokenize (tokenization relies to ending punctuation), then remove noise (removes punctuation, special characters, and white space)

```python
from nltk.tokenize import sent_tokenize

story = sent_tokenize(story)
story = [removing_noise(line) for line in story]
```

The last step is to tokenize the words in each sentence

```python
from nltk.tokenize import word_tokenize

story_w_tokens = [word_tokenize(sent) for sent in story]
# print(story[0])
# print(story_w_tokens[0])
```

### Normalization
Normalization transforms the text to a consistent, standard format. 


#### Stop Words
Removing stop words. Stop words are words like "the", "is", "and" which offer little information. They are like noise in natural language processing. 

```python
from nltk.corpus import stopwords 

stop_words = set(stopwords.words("english"))

story_no_stop = []
for sent in story_w_tokens:
    story_no_stop.append([word for word in sent if word not in stop_words])

# print(len(story_w_tokens[0]))
# print(len(story_no_stop[0]))
```

#### Stemming
Stemming reduces words to their base form by removing suffixes and prefixes, such as stemming "running", "run", and "runner" to "run". 

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

stemmed_story = []
for sent in story_no_stop:
    stemmed_story.append([stemmer.stem(word) for word in sent])

# print(story_no_stop[0])
# print(stemmed_story[0])
```

#### Lemmatization
Lemmatization is a text normalization technique that reduces words to their base or dictionary form (known as a lemma) using linguistic knowledge, such as a word’s part of speech and meaning e.g. lemmatization would reduce “running” to “run” and “better” to “good”. Since lemmatization requires the part of speech, it is a less efficient approach than stemming.

To add Lemmatization, we need to add parts-of-speech tagging.

```python
from nltk import pos_tag

pos_story = [pos_tag(sent) for sent in story_no_stop]

# print(pos_story)
```

We need to link the NLTK parts-of-speech to the WordNet tags in order to user WordNet's lemmatization.

```python
from nltk.corpus import wordnet

def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN  # Default to noun if unknown
```

Now we can lemmatize each word using WordNetLemmatizer.

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatized_story = []

for pos_sent in pos_story:
    lemmatized_story.append([
        lemmatizer.lemmatize(word, get_wordnet_pos(pos_tag))
        for word, pos_tag in pos_sent
])

# print(lemmatized_story[0])
```

Summary of the NLP pipeline so far: Raw Text -> Sentence Tokenization -> Lowercasing / Noise Removal -> Word Tokenization -> Stop Word Removal -> Lemmatization  OR  Stemming

## Vectorization
Next step is Vectorization. Computers cannot make sense of words, so every token needs a numeric value. There's a couple different ways to vectorize: Bag of Words,

### Bag of Words
Bag of Words assigns a numeric value to every word based on the frequency in the text, completely ignoring semantic meaning.

```python
from sklearn.feature_extraction.text import CountVectorizer
final_story = [" ".join(sent) for sent in lemmatized_story]
# Create the vectorizer
vectorizer = CountVectorizer()

# Fit and transform the text
bow_vectors = vectorizer.fit_transform(final_story)
# print(vectorizer.get_feature_names_out())
# print(bow_vectors.toarray())
```

### N-gram
An N-gram is a continuous group of N tokens. N-gram helps answer the question: "Which words show up together?" N-grams are helpful to capture short phrases which aren't able to be expressed in the individual words e.g. "reset password" vs the individual words "reset" and "password." N-grams do not understand semantics.

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create an N-gram vectorizer (bigrams in this example)
ngram_vectorizer = CountVectorizer(ngram_range=(2, 2))

# Fit and transform the text
ngram_vectors = ngram_vectorizer.fit_transform(final_story)

# Inspect learned features and vectors
# print(ngram_vectorizer.get_feature_names_out())
# print(ngram_vectors.toarray())
```

The ngram_range parameter controls the size of word sequences extracted: (1, 1) → Unigrams (Bag of Words), (2, 2) → Bigrams only, (1, 2) → Unigrams and bigrams, and (2, 3) → Bigrams and trigrams

N-grams help bridge the gap between rule-based chatbots that depend on key-words and intent classification bots that use machine learning. N-grams can distinguish the difference between "turn" and "turn on" or "customer support" and "customer".
